## 3. Variables

Variables let you store values and reuse them.

```python
width = 8
height = 5
area = width * height

print("width =", width)
print("height =", height)
print("area =", area)

name = "Alice"
favorite_number = 12

print(name)
print(favorite_number)
```

```python
# Mini math lab
minutes = 90
hours = minutes / 60
print("90 minutes is", hours, "hours")

celsius = 20
fahrenheit = (9 / 5) * celsius + 32
print("20 C is", fahrenheit, "F")
```

## 4. Conditionals

Conditionals help Python make decisions.

```python
number = 7

if number > 0:
    print("positive")
elif number < 0:
    print("negative")
else:
    print("zero")

if number % 2 == 0:
    print("even")
else:
    print("odd")
```

```python
score = 88

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

## 5. Loops

Loops repeat actions. They are one of the most powerful parts of programming.

```python
print("Counting from 1 to 10")
for i in range(1, 11):
    print(i)
```

```python
print("Even numbers from 2 to 20")
for i in range(2, 21, 2):
    print(i)
```

```python
total = 0
for i in range(1, 11):
    total = total + i

print("Sum from 1 to 10 =", total)
```

```python
countdown = 5
while countdown > 0:
    print(countdown)
    countdown = countdown - 1

print("Lift off into math mode")
```

## 6. Lists

Lists store many values together.

```python
numbers = [3, 5, 8, 13]

print(numbers)
print("First number:", numbers[0])
print("Last number:", numbers[-1])
print("How many numbers?", len(numbers))
```

```python
numbers = [3, 5, 8, 13]
total = 0

for n in numbers:
    total = total + n

average = total / len(numbers)
print("Sum =", total)
print("Average =", average)
```

## 7. Functions

Functions let you create your own reusable tools.

```python
def square(x):
    return x * x


def cube(x):
    return x * x * x


print(square(5))
print(cube(3))
```

```python
def area_triangle(base, height):
    return 0.5 * base * height


def is_even(n):
    return n % 2 == 0


print(area_triangle(10, 6))
print(is_even(14))
print(is_even(15))
```

## 8. Tiny Math Toolbox

Now combine a few ideas into a mini toolkit.

```python
def average_of_three(a, b, c):
    return (a + b + c) / 3


def distance(speed, time):
    return speed * time


def celsius_to_fahrenheit(c):
    return (9 / 5) * c + 32


print("average_of_three(4, 7, 10) =", average_of_three(4, 7, 10))
print("distance(12, 5) =", distance(12, 5))
print("celsius_to_fahrenheit(20) =", celsius_to_fahrenheit(20))
```
