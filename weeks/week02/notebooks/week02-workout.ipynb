## 1. Problem Solving in Small Steps

A good coding habit is to break a task into:

- input
- process
- output

Example:

- input: an `x` value
- process: compute `y = 2x + 1`
- output: show `(x, y)`

```python
x = 4
y = 2 * x + 1
print("x =", x)
print("y =", y)
print("point =", (x, y))
```

## 2. Coordinates

A graph is made from points. A point is just an ordered pair like `(x, y)`.

```python
x = -3
y = x ** 2
print((x, y))

x = 5
y = 2 * x + 1
print((x, y))
```

## 3. Tables of Values

Instead of one point, we usually want many points.

```python
print("x   y = 2x + 1")
for x in range(-5, 6):
    y = 2 * x + 1
    print(x, y)
```

```python
print("x   y = x^2")
for x in range(-5, 6):
    y = x ** 2
    print(x, y)
```

Notice how the line changes steadily, while the parabola changes more quickly as `x` gets larger.

## 4. Building Lists for Plotting

Most plots use a list of x-values and a list of y-values.

```python
x_values = list(range(-5, 6))
y_values = []

for x in x_values: ## This loop iterates through each x value in the list x_values.
    y_values.append(2 * x + 1) ## .append() is a method that adds the value 2 * x + 1 to the end of the list y_values for each x in x_values, effectively creating a list of y values corresponding to the x values based on the equation y = 2x + 1.

print("x_values =", x_values)
print("y_values =", y_values)
```

```python
x_values = list(range(-5, 6))
y_values = []

for x in x_values:
    y_values.append(x ** 2)

points = []
for i in range(len(x_values)): ## This loop iterates through the indices of the x_values list using range(len(x_values)).
    points.append((x_values[i], y_values[i]))

print(points)
```

## 5. Pattern Building with Loops

Loops can create visual patterns, not just calculations.

```python
for i in range(1, 6):
    print("*" * i)
```

```python
for x in range(1, 6):
    print(x, x ** 2, x ** 3)
```

## 6. Helper Functions

Functions make it easier to reuse equations.

```python
def line(x):
    return 2 * x + 1


def parabola(x):
    return x ** 2


def shifted_parabola(x):
    return (x - 2) ** 2 + 1


print(line(3))
print(parabola(3))
print(shifted_parabola(3))
```

```python
def build_points(func, x_values): 
    y_values = []
    for x in x_values:
        y_values.append(func(x)) ## This loop iterates through each x value in the list x_values, applies the function func to it, and appends the result to the y_values list. The function func can be any of the functions defined earlier (line, parabola, shifted_parabola), allowing us to build a list of y values corresponding to the x values based on the specified function.
    return y_values


x_values = list(range(-5, 6))
print(build_points(line, x_values))
print(build_points(parabola, x_values))
```

## 7. First Plot with Matplotlib

If `matplotlib` is installed, this will draw your first graph.

```python
try:
    import matplotlib.pyplot as plt ## This line attempts to import the pyplot module from the matplotlib library and assigns it the alias plt. If matplotlib is not installed, this will raise an ImportError, which is handled in the except block.
    has_matplotlib = True
    print("matplotlib is ready") 
except ImportError:
    has_matplotlib = False
    print("matplotlib is not installed yet")
    print("You can install it with: python3 -m pip install matplotlib")
```

```python
if has_matplotlib:
    x_values = list(range(-5, 6))
    y_values = build_points(line, x_values) ## This line calls the build_points function with the line function and the x_values list as arguments, which generates a list of y values corresponding to the x values based on the line function (y = 2x + 1). The resulting y_values list is then used for plotting.

    plt.plot(x_values, y_values, marker="o") ## This line creates a plot of the x_values and y_values using the plot function from matplotlib. The marker="o" argument specifies that each point on the line should be marked with a circle (o). This will help visualize the points corresponding to the line defined by the function line(x) = 2x + 1.
    plt.title("Line: y = 2x + 1")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()
```

```python
if has_matplotlib:
    x_values = list(range(-5, 6))

    plt.plot(x_values, build_points(line, x_values), marker="o", label="y = 2x + 1")
    plt.plot(x_values, build_points(parabola, x_values), marker="o", label="y = x^2")
    plt.plot(x_values, build_points(shifted_parabola, x_values), marker="o", label="y = (x - 2)^2 + 1")
    plt.axhline(0, color="gray", linewidth=1)
    plt.axvline(0, color="gray", linewidth=1)
    plt.title("Comparing Equations")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend() ## This line adds a legend to the plot, which will display the labels specified in the label argument of each plt.plot() call. The legend helps differentiate between the different lines representing the equations y = 2x + 1, y = x^2, and y = (x - 2)^2 + 1.
    plt.grid(True)
    plt.show()
```

## 8. A Rough Circle

A circle can also be built from points. Here we use cosine and sine.

```python
import math ## This line imports the math module, which provides access to mathematical functions and constants, such as pi, sine, cosine, etc. In this case, we will use math.pi to calculate the angle theta for generating points on a circle.

circle_x = []
circle_y = []
radius = 4

for i in range(121):
    theta = 2 * math.pi * i / 120 ## This line calculates the angle theta for each point on the circle. The loop iterates 121 times, which means it will generate points at angles ranging from 0 to 2 * math.pi (inclusive) in increments of 2 * math.pi / 120. This ensures that we get a smooth circle with points evenly spaced around it.
    circle_x.append(radius * math.cos(theta))
    circle_y.append(radius * math.sin(theta))

print(circle_x[:5]) ## This line prints the first 5 elements of the circle_x list, which contains the x coordinates of the points on the circle. The [:5] syntax is used to slice the list and retrieve only the first 5 elements.
print(circle_y[:5])
```

```python
if has_matplotlib:
    plt.plot(circle_x, circle_y)
    plt.title("A Rough Circle")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```
