## 1. Divisibility and Remainders

The `%` operator gives the remainder. If the remainder is 0, one number divides another.

```python
print(12 % 3)
print(17 % 5)

print(12 % 3 == 0)
print(17 % 5 == 0)
```

```python
def is_divisible(a, b):
    return a % b == 0


print(is_divisible(12, 3))
print(is_divisible(17, 5))
```

## 2. Factors

A factor of `n` is a positive integer that divides `n` with no remainder.

```python
def factors(n):
    result = []
    for i in range(1, n + 1):
        if n % i == 0:
            result.append(i)
    return result


print(factors(12))
print(factors(18))
```

```python
def factor_count(n):
    return len(factors(n))


print(factor_count(12))
print(factor_count(13))
```

## 3. Greatest Common Divisor

The gcd of two numbers is the largest number that divides both.

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return abs(a)


print(gcd(12, 18))
print(gcd(20, 30))
```

## 4. Least Common Multiple

The lcm of two numbers is the smallest positive number that both divide.

```python
def lcm(a, b):
    return abs(a * b) // gcd(a, b)


print(lcm(6, 8))
print(lcm(12, 18))
```

## 5. Prime Numbers

A prime number is a number greater than 1 with exactly two positive factors: 1 and itself.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


print(is_prime(2))
print(is_prime(11))
print(is_prime(15))
```

```python
def primes_up_to(n):
    result = []
    for value in range(2, n + 1):
        if is_prime(value):
            result.append(value)
    return result


print(primes_up_to(50))
```

## 6. Algebra by Testing Values

One beginner-friendly way to solve equations in code is to test values and see which ones work.

```python
def linear(x):
    return 2 * x + 3


print(linear(4))
```

```python
solutions = []
for x in range(-10, 11):
    if 2 * x + 3 == 11:
        solutions.append(x)

print(solutions)
```

```python
solutions = []
for x in range(-10, 11):
    if x ** 2 - 5 * x + 6 == 0:
        solutions.append(x)

print(solutions)
```

```python
if has_sympy:
    x = sp.symbols("x")
    print(sp.solve(sp.Eq(2 * x + 3, 11), x))
    print(sp.solve(sp.Eq(x ** 2 - 5 * x + 6, 0), x))
    print(sp.factor(x ** 2 - 5 * x + 6))
```
